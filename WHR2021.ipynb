```python
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Datasets/world-happiness-report-2021.csv')
df.head().dropna()
```

```python
df.drop(columns = ['Standard error of ladder score', 'upperwhisker', 'lowerwhisker','Explained by: Log GDP per capita', 'Explained by: Social support',
         'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices',
         'Explained by: Generosity', 'Explained by: Perceptions of corruption','Dystopia + residual'],inplace=True)
df
```

#### Ladder Score = The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you

```python
countries = df.head(6).append(df.tail(6))
```

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(y='Country name',x='Ladder score', data=countries)
```

### Plotting Ladder score w.r.t Regional indicator.

```python
plt.figure(figsize=(8,8))
sns.barplot(x='Ladder score', y='Regional indicator', data=df)
```

### Logged GDP per capita = Per capita gross domestic product (GDP) is a metric that breaks down a country's economic output per person and is calculated by dividing the GDP of a country by its population

```python
plt.figure(figsize=(8,8))
sns.boxplot(x='Logged GDP per capita', y='Regional indicator', data=df)
```

### Social Support

```python
plt.figure(figsize=(8,8))
sns.barplot(y='Regional indicator',x='Social support',data=df, palette='twilight_shifted', hatch='/')
```

### Mean of all the Features grouped by Regional Indicator

```python
abc = df.groupby('Regional indicator').mean()
abc
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(abc,cmap='twilight', center=0, annot=True, annot_kws={'fontsize': 15, 'fontweight': 'bold'}, linewidth=1, linecolor='black')
```

### Comparison of neighbouring countries with India in South Asia region

```python
data=df.loc[df['Regional indicator']=='South Asia'][:]
feature=['Ladder score',
       'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
plt.figure(figsize=(10,10))
for a in feature:
    plt.subplot(4,2,feature.index(a)+1)
    sns.barplot(y=data.drop(index=data[data['Country name']=='India'].index)['Country name'],x=data[a], palette='viridis')
    plt.subplot(4,2,feature.index(a)+1).axvline(data.loc[data['Country name']=='India'][a].values,ls='dotted',color='black',label="India's score", lw=5)
plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05,5.5))
plt.show()
```

### Establishing Correlation matrix 

```python
pq = df.drop(columns=['Ladder score in Dystopia']).corr()
pq
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(pq, cmap='cividis', center=0, annot=True, annot_kws={'fontsize': 12, 'fontweight': 'bold'}, linewidth=1, linecolor='black', square=True) 
```
